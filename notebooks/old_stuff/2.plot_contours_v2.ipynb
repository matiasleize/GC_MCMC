```python
labels = [r'$M_{abs}$',r'$\beta$',r'$H_{0}$',r'$\Omega_m^{\rm \Lambda CDM}$']
names = [i.replace('$','') for i in labels]; 

labels_filt = [r'$\beta$',r'$H_{0}$']
names_filt = [i.replace('$','') for i in labels_filt]; 


filename_1 = home_directory+'/Documents/Repos/GILA-output/L=0.75/sample_GILA_PPS_CC_BAO_4params/sample_GILA_PPS_CC_BAO_4params'
filename_2 = home_directory+'/Documents/Repos/GILA-output/L=0.80/sample_GILA_PPS_CC_BAO_4params/sample_GILA_PPS_CC_BAO_4params'
filename_3 = home_directory+'/Documents/Repos/GILA-output/L=0.85/sample_GILA_PPS_CC_BAO_4params/sample_GILA_PPS_CC_BAO_4params'
filename_4 = home_directory+'/Documents/Repos/GILA-output/L=0.90/sample_GILA_PPS_CC_BAO_4params/sample_GILA_PPS_CC_BAO_4params'
filename_5 = home_directory+'/Documents/Repos/GILA-output/L=1.00/sample_GILA_PPS_CC_BAO_4params/sample_GILA_PPS_CC_BAO_4params'


with np.load(filename_1 + '_deriv.npz') as data:
    samples_1 = data['new_samples']

with np.load(filename_2 + '_deriv.npz') as data:
    samples_2 = data['new_samples']

with np.load(filename_3 + '_deriv.npz') as data:
    samples_3 = data['new_samples']

with np.load(filename_4 + '_deriv.npz') as data:
    samples_4 = data['new_samples']

with np.load(filename_5 + '_deriv.npz') as data:
    samples_5 = data['new_samples']


samps_1 = MCSamples(samples = samples_1,names=names,labels=names)
samps_2 = MCSamples(samples = samples_2,names=names,labels=names)
samps_3 = MCSamples(samples = samples_3,names=names,labels=names)
samps_4 = MCSamples(samples = samples_4,names=names,labels=names)
samps_5 = MCSamples(samples = samples_5,names=names,labels=names)

samps_1 = samps_1.copy(label=r'Lowest-order with $0.3\sigma$ smoothing',
            settings={'mult_bias_correction_order':0,'smooth_scale_2D':0.3,
            'smooth_scale_1D':0.3})

samps_2 = samps_2.copy(label=r'Lowest-order with $0.3\sigma$ smoothing',
            settings={'mult_bias_correction_order':0,'smooth_scale_2D':0.3,
            'smooth_scale_1D':0.3})

samps_3 = samps_3.copy(label=r'Lowest-order with $0.3\sigma$ smoothing',
            settings={'mult_bias_correction_order':0,'smooth_scale_2D':0.3,
            'smooth_scale_1D':0.3})

samps_4 = samps_4.copy(label=r'Lowest-order with $0.3\sigma$ smoothing',
            settings={'mult_bias_correction_order':0,'smooth_scale_2D':0.3,
            'smooth_scale_1D':0.3})

samps_5 = samps_5.copy(label=r'Lowest-order with $0.3\sigma$ smoothing',
            settings={'mult_bias_correction_order':0,'smooth_scale_2D':0.3,
            'smooth_scale_1D':0.3})
print(samps_1.paramNames)

#print(samps_1, samps_2)
#print(type(samps_1))
g = plots.get_subplot_plotter()
g.triangle_plot([samps_1, samps_2, samps_3, samps_4, samps_5], filled=True, params=names_filt, param_names=names_filt, legend_labels=['L=0.75','L=0.80','L=0.85','L=0.90','L=1.00'],
                 param_limits={r'\beta':[0,12],'H_{0}':[68,82]}) 

plt.savefig('beta_H0.pdf')

```
